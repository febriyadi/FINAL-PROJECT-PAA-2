Bubble & Insertion Sort


```python
import time

def bubble_sort(arr):
    n = len(arr)
    start_time = time.time()

    for i in range(n):
        swapped = False
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break

    end_time = time.time()
    execution_time = end_time - start_time
    return arr, execution_time

def insertion_sort(arr):
    n = len(arr)
    start_time = time.time()

    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

    end_time = time.time()
    execution_time = end_time - start_time
    return arr, execution_time

def print_iterations(arr, sort_type):
    if sort_type == "bubble":
        bubble_sort(arr)
    elif sort_type == "insertion":
        insertion_sort(arr)

    print(f"5 iterasi pertama {sort_type}:")
    for i in range(min(5, len(arr))):
        print(arr[i], end=" ")
    print("\n")

    print(f"5 iterasi terakhir {sort_type}:")
    for i in range(max(0, len(arr)-5), len(arr)):
        print(arr[i], end=" ")
    print("\n")

def print_before_after(arr, sorted_arr):
    print("\nSebelum pengurutan:", arr)
    print("Setelah pengurutan:", sorted_arr)

def main():
    arr = [12, 99, 62, 15, 20, 95, 39, 48, 3, 24, 8, 43, 74, 19, 97, 33, 49, 68, 55, 33, 90, 90, 7,
           26, 85, 46, 39, 40, 9, 36, 60, 64, 89, 31, 25, 71, 21, 23, 63, 84, 32, 5, 3, 44, 21, 10, 21,
           17, 50, 2, 36, 53, 79, 54, 19, 88, 1, 32, 31, 15, 6, 3, 1, 40, 22, 43, 18, 1, 77, 9, 59,
           40, 7, 41, 81]

    print("\nProgram Pengurutan Menggunakan Bubble Sort dan Insertion Sort")
    print()
    print("Data Yang Akan Diurutkan")
    print(arr)
    print()

    choice = input("Pilih algoritma pengurutan (bubble atau insertion): ")
    if choice == "bubble":
        sorted_arr, execution_time = bubble_sort(arr.copy())
        print_iterations(arr, "Bubble Sort")
        print(f"Waktu eksekusi:", "{:.4f}".format(execution_time), "detik")
        print_before_after(arr.copy(), sorted_arr)
    elif choice == "insertion":
        sorted_arr, execution_time = insertion_sort(arr.copy())
        print_iterations(arr,"Insertion Sort")
        print(f"Waktu eksekusi:", "{:.4f}".format(execution_time), "detik")
        print_before_after(arr.copy(), sorted_arr)
    else:
        print("Pilihan tidak valid")
        main()
        return
    analyze_again = input("\nApakah Anda ingin menganalisis algoritma lainnya? (ya/tidak): ")
    if analyze_again.lower() == "ya":
        main()
    else:
        print("Terima kasih telah menggunakan program ini. Sampai jumpa!")

# Program Utama
main()
```

Algoritma TSP dan Djikstra

```python
import time
import itertools
from heapq import heappop, heappush

# Representasi grafik sebagai dictionary
graph = {
    'a': {'b': 12, 'c': 10, 'g': 12},
    'b': {'a': 12, 'c': 8, 'd': 12},
    'c': {'a': 10, 'b': 8, 'd': 11, 'e': 3, 'g': 9},
    'd': {'b': 12, 'c': 11, 'e': 11, 'f': 10},
    'e': {'c': 3, 'd': 11, 'f': 6, 'g': 7},
    'f': {'d': 10, 'e': 6, 'g': 9},
    'g': {'a': 12, 'c': 9, 'e': 7, 'f': 9}
}

def tsp(graph, source, destination):
    start_time = time.time()

    # Menghasilkan semua kemungkinan permutasi
    nodes = list(graph.keys())
    permutations = list(itertools.permutations(nodes))

    shortest_path = None
    shortest_distance = float('inf')

    iteration_count = 0

    for path in permutations:
        if path[0] == source and path[-1] == destination:
            distance = 0
            valid_path = True
            for i in range(len(path) - 1):
                current_node = path[i]
                next_node = path[i + 1]

                if next_node not in graph[current_node]:
                    valid_path = False
                    break

                distance += graph[current_node][next_node]

            if valid_path and distance < shortest_distance:
                shortest_distance = distance
                shortest_path = path

            iteration_count += 1
            print(f"Iterasi ke-{iteration_count}: {path} - Jarak: {distance}")

    end_time = time.time()
    computation_time = end_time - start_time

    print("\nHasil Akhir:")
    print(f"Jalur Terpendek: {shortest_path}")
    print(f"Jarak Terpendek: {shortest_distance}")
    print(f"Waktu Komputasi: {computation_time} detik")

def dijkstra(graph, source, destination):
    start_time = time.time()

    # Inisialisasi jarak awal semua node dengan infinity
    distances = {node: float('inf') for node in graph}
    distances[source] = 0

    # Heap berisi node-node yang akan diproses
    heap = [(0, source)]

    iteration_count = 0

    while heap:
        current_distance, current_node = heappop(heap)

        if current_distance > distances[current_node]:
            continue

        if current_node == destination:
            break

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heappush(heap, (distance, neighbor))

        iteration_count += 1
        print(f"Iterasi ke-{iteration_count}: Node: {current_node}, Jarak: {current_distance}")

    end_time = time.time()
    computation_time = end_time - start_time

    print("\nHasil Akhir:")
    print(f"Jarak Terpendek dari node {source} ke node {destination}: {distances[destination]}")
    print(f"Waktu Komputasi: {computation_time} detik")

# Main Program
while True:
    print("\nPilih algoritma:")
    print("1. TSP")
    print("2. Djikstra")
    print("3. Keluar")
    user_choice = input("Masukkan pilihan: ")

    if user_choice == "1":
        source = input("Masukkan node awal (source): ")
        destination = input("Masukkan node tujuan (destination): ")
        tsp(graph, source, destination)
    elif user_choice == "2":
        source = input("Masukkan node awal (source): ")
        destination = input("Masukkan node tujuan (destination): ")
        dijkstra(graph, source, destination)
    elif user_choice == "3":
        break
    else:
        print("Pilihan tidak valid. Silakan pilih 1, 2, atau 3.")
```
